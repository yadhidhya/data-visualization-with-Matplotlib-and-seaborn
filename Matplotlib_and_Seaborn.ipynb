**Line chart**

Line charts are one of the simplest and most widely used data visualization techniques. A line chart displays information as a series of data points or markers, connected by straight lines. You can coustomize the shape, size, color and other aesthetic elements of the markers and lines for better visual clarity

Here's python list showing the yield of apples(tons per hectare) over 6 years in an imaginary country called kanto


```python
yield_apples=[0.854,0.92,0.890,0.987,0.889,0.971]
```

we can visualize how the yield of apples changes over time using a line chart. To draw a line chart, we can use **plt.plot** function

```python
plt.plot(yield_apples)
```

calling the plt.plot function draws the line chart as expected, and also returns a list of plots drawn [<matplotlib.lines.Line2D at 0x7dbb39b99290>] shown within the output. We can include a semicolon  ( ; ) at the end of the statement in the cell to avoiding showing the output and just display the graphs

```python
plt.plot(yield_apples);
```

Let's enhance this plot step-by-step to make it more informative and beautiful

**Coustomizing the X-axis**

The X-axis of the plot currently shows list elements indexes 0 to 5. The plot would be more informative if we could show the year for which the data is being plotted. We can do this by two arguments plt.plot.

```python
years=[2010,2011,2012,2013,2014,2015]
yield_apples=[0.854,0.92,0.890,0.987,0.889,0.971]
plt.plot(years,yield_apples);
```

**Axis Labels**

We can add labels(names) to the axes to show what each axis represents using the plt.xlabel and plt.ylabel methods

```python
plt.xlabel("year")
plt.ylabel("yield(tons per hectare)")
plt.plot(years,yield_apples);
```

```python
oranges =[0.862,0.88,0.92,0.931,0.945,0.971]
plt.plot(years,yield_apples)
plt.plot(years,oranges)
plt.xlabel("year")
plt.ylabel("yield(tons per hectare)")
```

**Chart Title and Legend**

To differentiate between multiple lines, we can include a legend within the graph using plt.legend function. We also give the entire chart a title using plt.title function

```python
plt.plot(years,yield_apples)
plt.plot(years,oranges)
plt.xlabel("year")
plt.ylabel("yield(tons per hectare)")
plt.legend(['yield_apples','oranges'])
plt.title("yield of apples and oranges");
```

**Line Markers**

We can also show markers for the data points on each line using marker argument of plt.plot. Matplotlib supports many different types of markers like circle, cross, square, diamond etc. You can find the full list of marker types here:

https://matplotlib.org/3.1.1/api/markers_api.html

```python
plt.plot(years,yield_apples,marker='o')
plt.plot(years,oranges,marker='*')
plt.xlabel("year")
plt.ylabel("yield(tons per hectare)")
plt.legend(['yield_apples','oranges'])
plt.title("yield of apples and oranges");
```

**Styling lines and markers**

The plt.plot function supports many arguments for styling lines and markers:



*   color or c: set the color of the line.

*   linestyle or ls: choose between a solid or dashed line.

*   linewidth or lw: set the width of a line.

*   markersize or ms: set the size of markers.
*   markeredgecolor or mec: set the edge color for markers.


*   markeredgewidth or mew: set the edge width for markers.


*   markerfacecolor or mfc: set the fill color for markers.


*   alpha: opacity of the plot

check out the documentation for plt.plot to learn more:

https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot

List of colors supported by matplotlib:

https://matplotlib.org/stable/gallery/color/named_colors.html

**Let's see the example of styling the lines and markers**



```python
plt.plot(years, yield_apples, marker='o', c='b', ls='-',lw=2, ms=8,mew=2,mec='navy')
plt.plot(years, oranges, marker='*', c='r', ls='--',lw=2, ms=8,mew=2,mec='g',mfc='w', alpha=.5)
plt.xlabel("year")
plt.ylabel("yield(tons per hectare)")
plt.legend(['yield_apples','oranges'])
plt.title("yield of apples and oranges");
```

**Note**

keep semicolumn at the end of the code so that you cannot get the extra information while calling the plt.plot function

**The fmt argument provides a shorthand for specifying the line style, marker and line color. It can be provided as the third argument to plt.plot**

```python
plt.plot(years,yield_apples,'o:b')
plt.plot(years,oranges,'*--r')
plt.xlabel("year")
plt.ylabel("yield(tons per hectare)")
plt.legend(['apples','oranges'])
plt.title("yield of apples and oranges");
```

If no line style is specified in fmt, only markers are drawn.

**Scatter Plot**

In a scatter plot, the values of 2 variables are plotted as points on a 2-dimensional grid. Additionally, you can also use a third variable to determine the size or color of the points. Let's try out an example.

The Iris flower dataset provides samples measurements of  sepals and petals for 3 species of flowers. The iris dataset is included with the seaborn library, and can be loaded as a pandas dataframe. So there is no need to download a dataframe it is inbuiltly available

```python
#load data into pandas dataframe
flowers_df = sns.load_dataset("iris")
```

```python
flowers_df
```

```python
flowers_df.species.unique()
```

Let's try to visualize the relationship between sepal length and sepal width. Our first instinct might be to create line chart using plt.plot. However, the output is not very informative as there are too many combinations of the two properties within the dataset, and there doesn't seem to be simple relationship between them.

```python
plt.plot(flowers_df.sepal_length,flowers_df.sepal_width);
```

You can see that we cannot understand anything by visualizing the data in line chart so let's use scatterplot function from seaborn (imported as sns).

```python
import seaborn as sns
sns.scatterplot(x=flowers_df.sepal_length, y=flowers_df.sepal_width);
```

Matplotlib can also allows you to draw the scatterplots but not powerful as seaborn. so, seaborn is much nicer to use

```python
#changing the facecolor of the graph
import matplotlib
matplotlib.rcParams['figure.facecolor'] = '#000000'
```

```python
sns.scatterplot(x=flowers_df.sepal_length, y=flowers_df.sepal_width);
```

```python
matplotlib.rcParams['figure.facecolor'] = 'white'
sns.scatterplot(x=flowers_df.sepal_length, y=flowers_df.sepal_width);
```

You can observe that seaborn automatically given labels to X and Y axis which makes it different from matplotlib.

**Adding Hues**

Notice how the points in the above plot seem to form distinct clusters with some outliers. We can color the dots using the flower species as a "hue". We can also make the points larger using the "s" argument

```python
sns.scatterplot(x=flowers_df.sepal_length,
                y=flowers_df.sepal_width,
                hue=flowers_df.species, s=100);
```

Adding hues makes the plot more informative. We can immediately tell that flowers of the setosa species have a smaller sepal length but higher sepal widths, while the opposite holds true for the viginica species.

**Customizing Seaborn Figures**

Since Seaborn uses Matplotlib's plotting functions internally, we can use functions like plt.figure and plt.title to modify the figure

```python
plt.figure(figsize=(12,6))
plt.title('sepal Dimensions')

sns.scatterplot(x=flowers_df.sepal_length,
                y=flowers_df.sepal_width,
                hue=flowers_df.species, s=100);
```

**Plotting using Pandas Data Frames**

Seaborn has in-built support for Pandas data frames. Instead of passing each column as a series, you can also pass column names and use the "data" argument to pass the data frame.

```python
plt.title('sepal Dimensions')

sns.scatterplot(x='sepal_length',
                y='sepal_width',
                hue='species', s=100, data=flowers_df);
```

**Histogram**



A histogram represents the distribution of data by forming bins along the range of the data and then drawing bars to show the number of observations that fall in each bin

As an example, let's visualize how the values of sepal width in the flowers dataset are distributed. We can use the "plt.hist" function to create a histogram.

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
flowers_df=sns.load_dataset("iris")
plt.title("Distribution of sepal width")
plt.hist(flowers_df.sepal_width);
```

```python
sns.set_style("darkgrid")
```

```python
flowers_df=sns.load_dataset("iris")
plt.title("Distribution of sepal width")
plt.hist(flowers_df.sepal_width);
```

here we can observe that the histogram is divided into 10 parts, it is deafault and for every histogram, it is same and we can change that

the bars in the histogram are called as bins or intervals

let's change the number of bins to 5 for clear visualization.

```python
plt.hist(flowers_df.sepal_width,bins=5);
```

We can specify that at what points on x-axis the bins should be created

```python
import numpy as np
np.arange(2, 5, 0.25)
```

```python
import numpy as np

# Specifying the boundaries of each bin
plt.hist(flowers_df.sepal_width, bins=np.arange(2, 5, 0.25));
```

```python
# Bins of unequal sizes
plt.hist(flowers_df.sepal_width, bins=[1, 3, 4, 4.5]);
```

**Multiple Histograms**

Similar to line charts, we can draw multiple histograms in a single chart. We can reduce the opacity of each histogram, so the bars of one histogram don't hide the bars for others.

Let's draw separate histograms for each species of flowers.

```python
setosa_df = flowers_df[flowers_df.species == 'setosa']
versicolor_df = flowers_df[flowers_df.species == 'versicolor']
virginica_df = flowers_df[flowers_df.species == 'virginica']
```

```python
plt.hist(setosa_df.sepal_width, alpha=0.4, bins=np.arange(2, 5, 0.25));
plt.hist(versicolor_df.sepal_width, alpha=0.4, bins=np.arange(2, 5, 0.25));

plt.legend(['setosa', 'versicolor']);
```

We can also stack multiple histograms on the top of one another

```python
plt.title("distribution of sepal width")
plt.hist([setosa_df.sepal_width, versicolor_df.sepal_width, virginica_df.sepal_width],
         bins=np.arange(2,5,0.25),stacked=True);
plt.legend(['setosa', 'versicolor', 'virginica']);
```

**Bar Chart**

Bar charts are quite similar to line charts i.e. they show a sequence of values, however a bar is shown for each value, rather than points connected by lines. We can use the plt.bar function to draw a bar chart

```python
years=range(2000,2006)
apples=[0.35,0.6,0.9,0.8,0.65,0.8]
oranges=[0.4,0.8,0.9,0.7,0.6,0.8]
```

```python
#line chart
plt.plot(years,apples);
```

```python
#bar chart
plt.bar(years,apples);
```

You can draw bar charts and line charts together

```python
plt.plot(years,apples ,'o--r')
plt.bar(years,apples);
plt.title("yeild of apples");
```

Like histograms, bars can also be stacked on top of one another. We use the bottom argument to plt.bar to achieve this.

```python
plt.bar(years, apples, bottom=oranges);
plt.bar(years, oranges);
plt.legend(['apples', 'oranges']);
```

**Bar Plots with Averages**

Let's look at another sample dataset included with Seaborn, called "tips". The dataset contains information about sex, time of day, total bill and tip amount for customers visiting a restuarant over a week.

```python
tips_df=sns.load_dataset("tips");
```

```python
tips_df
```

We might want to draw a bar chart to visualize how the average bill amount varies across different days of the week. One way to do this would be to compute the day-wise averages using **groupby** and then use plt.bar

```python
bill_avg=tips_df.groupby('day')[['total_bill']].mean()
plt.bar(bill_avg.index,bill_avg.total_bill);
```

Using Seaborn, it is easy to compute average by using inbuilt function "sns.barplot" so no need to use groupby function

```python
sns.barplot(x='day',y='total_bill',data=tips_df);
```

The lines cutting each bar represent the amount of variations in the values. For instance, it seems like the variation in the total bills was quite high on Fridays, and lower on Saturday.

We can also specify a hue argument to compare bar plots side-by-side based on a third feature e.g. sex

```python
sns.barplot(x='day',y='total_bill',hue='sex',data = tips_df);
```

You can make the bars horizontal simply by switching the axes

```python
sns.barplot(x='total_bill',y='day',hue='sex',data = tips_df);
```

**Heat Map**

A heatmap is used to visualize 2-dimensional data like a matrix or a table using colors. The best way to understand it by looking at an example. We'll use another sample dataset from Seaborn, called "flights", to visualize monthly passenger footfall at an airport over 12 years

```python
flights_df=sns.load_dataset("flights")

```

```python
flights_df
```

```python
plt.plot(flights_df.passengers);
```

```python
flights_df=sns.load_dataset("flights").pivot(index="month", columns="year", values="passengers")
```

```python
flights_df
```

flights_df is a matrix with one row for each month and one column of each year. The values in the matrix show the number of passengers(in thousands) that visited the airport in a specific month of a specific year. We can use the sns.heatmap function to visualize the footfall at the airport

```python
plt.title("No. of passengers in thousands")
sns.heatmap(flights_df);
```

The brighter colors indicate a higher footfall at the airport. By looking at the graph, we can infer two things:



1.   The footfall at the airport in any given year tends to be the higher around july and august.
2.   The footfall at the airport in any given month tends to grow year by year.

We can also display the actual values (above heatmap doesn't contains values it contains only colors) in each block by specifying annot=True, and use the cmap argument to change the color palette.



```python
plt.title("No. of passengers in thousands")
sns.heatmap(flights_df, fmt='d', annot=True,cmap='Blues');
```

**Plotting Multiple charts in a grid**

Matplotlib and Seaborn also support plotting multiple charts in a grid, using plt.subplots, which returns a set of axes that can be used for plotting.

Here's a single grid showing different types of charts we've covered in this notebook.

```python
plt.subplots(2,3)
```

```python
plt.subplots(2,3,figsize=(12,8))
```

```python
#spacing between each graph
plt.subplots(2,3,figsize=(12,8))
plt.tight_layout(pad=3)
```

```python
fig, axes=plt.subplots(2,3,figsize=(12,8))

# line chart
axes[0,0].plot(years,apples,'o--r')
axes[0,0].plot(years,oranges,'*--b')
# use set_title instead of title
axes[0,0].set_title('crop yields')
axes[0,0].set_xlabel('year')
axes[0,0].set_ylabel('yield(tons per hectare)')
axes[0,0].legend(['apples', 'oranges']);

#scatter plot
axes[0,1].set_title('sepal length vs. width')
sns.scatterplot(x=flowers_df.sepal_length,
                y=flowers_df.sepal_width,
                hue=flowers_df.species, s=100, ax=axes[0,1]);

#histogram
axes[0,2].set_title('distribution of sepal width')
axes[0,2].hist(flowers_df.sepal_width,bins=np.arange(2,5,0.5));

#bargraph
axes[1,0].set_title('yeild of apples')
axes[1,0].bar(years,apples);
axes[1,0].bar(years,oranges,bottom=apples);
axes[1,0].legend(['apples', 'oranges']);

#heatmap
axes[1,1].set_title('No. of passengers in thousands')
sns.heatmap(flights_df, fmt='d', cmap='Blues',ax=axes[1,1]);

#image
axes[1,2].set_title('image')
axes[1,2].imshow(img)
axes[1,2].grid(False)
axes[1,2].axis('off')

plt.tight_layout(pad=2);

```
